```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Trigonometric functions

x = np.sin(np.pi/2)
y = np.cos(np.pi/2)
z = np.tan(np.pi/2)

print(x)
print(y)
print(z)
```

```python
# array of trigonometric functions

x = np.array([np.pi/2, np.pi/3, np.pi/4, np.pi/5])

y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

print(y_sin)
print(y_cos)
print(y_tan)

# Create a plot with multiple subplots
plt.figure(figsize=(10, 6))

# Plot Sine
plt.subplot(3, 1, 1)
plt.plot(x, y_sin, label="sin(x)", color="blue")
plt.title("Sine Function")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.grid(True)
plt.legend()

# Plot Cosine
plt.subplot(3, 1, 2)
plt.plot(x, y_cos, label="cos(x)", color="green")
plt.title("Cosine Function")
plt.xlabel("x (radians)")
plt.ylabel("cos(x)")
plt.grid(True)
plt.legend()

# Plot Tangent (with some limits to avoid extreme values)
plt.subplot(3, 1, 3)
plt.plot(x, y_tan, label="tan(x)", color="red")
plt.title("Tangent Function")
plt.xlabel("x (radians)")
plt.ylabel("tan(x)")
plt.ylim(-10, 10)  # Limit the y-axis to avoid large values near π/2 and 3π/2
plt.grid(True)
plt.legend()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
x = np.arange(0, 3 * np.pi, 0.1) # create x array of angels from range 0 to 3*3.14
y = np.sin(x) # create y array of sine values of angles from x array

plt.plot(x, y) # plot grah
plt.title(" Graphical Representation of sine function")
plt.xlabel("x axis ")
plt.ylabel("y axis ")
plt.show() # show plotted graph
```
